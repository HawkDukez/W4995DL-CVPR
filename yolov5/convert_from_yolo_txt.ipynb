```python
from PIL import Image
import os,sys

image_w = 1845
image_h = 750

PATH = "/home/pz2272/yolov5/runs/detect/exp/"
OUTPUT_PATH = "/home/pz2272/yolov5/submit2/"

images_dir = os.listdir(PATH)
images_dir.sort()
index = -1
for i in range(len(images_dir)):
    if images_dir[i] == "labels":
        index = i
    else:
        images_dir[i] = images_dir[i].replace(".jpg",".txt")
images_dir.pop(index)
images_dir = [os.path.join(OUTPUT_PATH,name) for name in images_dir]

for file_dir in images_dir:
     with open(file_dir,"x") as f:
            f.close()
    
labels_dir = os.listdir(PATH+"labels")
labels = [(os.path.join(PATH,"labels",label),label) for label in labels_dir]

for lab in labels:
    label_name = lab[1]
    with open(lab[0],"r") as f:
        content = f.read().split("\n")
        output = ""
        for item in content[:-1]:
            temp = item.split(" ")

            center_x = float(temp[1])*image_w 
            center_y = float(temp[2])*image_h 
            width = float(temp[3])*image_w
            height = float(temp[4])*image_h 
            conf = float(temp[5])
            
            x_min = center_x - width/2
            y_min = center_y - height/2
            x_max = center_x + width/2
            y_max = center_y + height/2
            
            output+= "{} {:.2f} {:.2f} {:.2f} {:.3f} {:.3f}\n".format("vehicle", x_min, y_min, x_max, y_max,conf)
            
        with open(os.path.join(OUTPUT_PATH,label_name),"w") as f:
            f.write(output)  

for file_dir in images_dir:
     with open(file_dir,"r+") as f:
            content = f.read()
            if content == "":
                f.write("{} {:.2f} {:.2f} {:.2f} {:.3f} {:.3f}\n".format("vehicle", 0, 0, 0, 0, 0))  

```
